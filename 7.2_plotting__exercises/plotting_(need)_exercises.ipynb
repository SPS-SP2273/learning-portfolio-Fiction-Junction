```python
import matplotlib.pyplot as plt
import numpy as np
```

### Exercise 1
Fill in the blanks(indicated with a ?), so the code below will produce the plot shown.

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='red', linestyle='dashed',
             marker='o', markerfacecolor='yellow',
             ecolor='black',capsize=5,
             label='Y values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right')
plt.show()
```

### Exercise 2
1. Use NumPy to generate $x$ and $y$ data suitable for plotting the graph of the function $$ y = e^{-x/10}\sin(x) $$ for values of $x$ in the range 0 to 50.

2. Customise your plot so that:
    * $x$ and $y$ axes have labels of fontsize 15
    * $x$ and $y$ labels are in the color darkorange
    * Grid with opacity of 25%
    * Indicate equation of function in title with font size 20
    * Change $y$ limit using ylim to [-1,1]

```python
x = np.linspace(0, 50, 200)
y = np.exp(-x/10) * np.sin(x)

plt.plot(x, y, color='blue')
plt.xlabel('x', fontsize=15, color='darkorange')
plt.ylabel('y', fontsize=15, color='darkorange', rotation=0)
plt.title(r"$y=e^{-x/10}\sin(x)$", fontsize=20)
plt.grid(alpha=0.25)
plt.ylim(-1,1)
plt.show()
```
